```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#import dataset

dataset = pd.read_csv('Salary_Data.csv')
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,1].values

# dataset

```

```python
# x
```

```python
# y
```

```python
#splitting the dataset into the training and testing 
# !pip install sklearn
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=1/3,random_state = 0)
```

```python
# x_train
# y_train

# x_test
# y_test
```

```python
#fitting line

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()

regressor.fit(x_train,y_train)
```

```python
#predicting train set result
y_trainpred = regressor.predict(x_train)
#predicting test set Result
y_testpred = regressor.predict(x_test)
```

```python
#visualizing data in graph for train set
plt.scatter(x_train,y_train,color='green')
plt.plot(x_train,regressor.predict(x_train),color='blue')
plt.title('salary vs years of expe[for training set]')
plt.xlabel('Years Experience')
plt.ylabel('Salary')
plt.show()
```

```python
#visualizing data in graph for test set
plt.scatter(x_test,y_test,color='green')
plt.plot(x_test,regressor.predict(x_test),color='blue')
plt.title('salary vs years of expe[for testing set]')
plt.xlabel('Years Experience')
plt.ylabel('Salary')
plt.show()
```
